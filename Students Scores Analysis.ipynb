```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("students_scores.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0", axis = 1)
print(df.head())
```

```python
plt.figure(figsize=(5,5))
plt.title ("Gender Distribution")
ax = sns.countplot(data=df, x="Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the male
```

```python
gb = df.groupby("WklyStudyHours").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb)
plt.title ("Study Hours vs Scores")
plt.show()
```

#from the above chart we have concluded that the who study more have good impact on there scores.

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb)
plt.title ("Relationship Between Parent's Education and Student's Score")
plt.show()
```

#from the above chart we have concluded that the education of parents have good impact on there scores.

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(gb)
plt.title ("Relationship Between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/nigligible impact on the
#student's score due to their Parent's marital status
```

Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()

l= ["group A", "group B", "group C", "group D", "group E",]
mlist= [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
colors = ["#e6194b","#ffe119","#3cb44b","#4363d8","#f58231"]
plt.pie(mlist, labels = l, colors = colors, autopct = "%1.2f%%" )
plt.title ("Distribution of Ethnic Groups") 
plt.show()
```
